```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, kruskal

df = pd.read_csv("../outputs/features.csv")
```

Correlation

```python
for col in df.columns:
    if col not in ["AgatstonScore", "Category", "PatientID"]:
        corr, p = spearmanr(df[col], df["AgatstonScore"])
        print(col, corr, p)
```

Kruskal

```python
for col in df.columns:
    if col not in ["AgatstonScore", "Category", "PatientID"]:
        groups = [df[df["Category"] == i][col] for i in range(4)]
        stat, p = kruskal(*groups)
        print(col, p)
```

Heatmap

```python
sns.heatmap(df.corr(), annot=False)
plt.show()
```
